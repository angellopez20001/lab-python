### Reduction operation: the sum of the numbers in the range [0, a)

```python
import time

def reduc_operation(a):
    """Compute the sum of the numbers in the range [0, a)."""
    x = 0
    for i in range(a):
        x += i
    return x

# Secuencial

value = 1000000

initialTime = time.time()
suma = reduc_operation(value)
finalTime = time.time()

print("Time taken by reduction operation:", (finalTime - initialTime), "seconds")

# Utilizando las operaciones mágicas de ipython
%timeit -r 2 reduc_operation(value)

print(f"\n \t Computing the sum of numbers in the range [0, value): {suma}\n")
```

## Ejercicio 3.2 Apartado A

```python
import time

def sumacomponentesdelista(a):
    """Compute the sum of the numbers in the range [0, a)."""
    x = 0
    for i in a:
        x += i
    return x

    
listaasumar=list(range(1000000))
##Primero usamos el bucle for
initialtime = time.time()
suma= sumacomponentesdelista(listaasumar)
finaltime=time.time()
tiempotot=finaltime-initialtime

print ("El numero total sumado es de "+str(suma))
print("El tiempo de ejecucion de la suma de la lista por bucle for es de:" +" " + str(tiempotot)+ " " + "segundos")
%timeit -r 2 sumacomponentesdelista(listaasumar)
##Ahora probamos con la funcion sum
initialtime = time.time()
suma= sum(listaasumar)
finaltime=time.time()
tiempotot=finaltime-initialtime

print("El tiempo de ejecucion de la suma de la lista por funcion sum es de:" +" " + str(tiempotot)+ " " + "segundos")
%timeit -r 2 sum(listaasumar)
```
